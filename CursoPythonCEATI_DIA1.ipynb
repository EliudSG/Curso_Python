# Operaciones Matematicas

```python
# 4to Nivel de importancia:
# Suma Ordinaria
print(3 + 4)
variable1, variable2 = 20,40 # tambien puedes ponerlas por separadas
print( variable1 + variable2 )

# Resta Ordinaria
print (2 - 5)
# Tambien podemos asignar variables
v1 = 10.4
v2 = 15
resultado = v2-v1
print(resultado)
```

```python
# 3er Nivel de importancia:
# Multiplicacion Ordinaria
M1, M2 = 2 , 3.14
resultado = M1 * M2
print( resultado )
# Division Ordinaria
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5

print(M1/M2) # 2/5 = 0.4
print(M3/M4) #6/5 = 1.2
print(M5/M6) # 0.8
# Division de Piso
print(M1 // M2) # Quiero que vean que la division de piso siempre redondea hacia abajo
# sin importar que tan cercano esté del redondeo hacia arriba.
print(M3 // M4)
print(M5 // M6)

print( 6 / 2 * 3 ) # Resultado del meme
```

```python
# Modulo (residuo)
# Si el cosiente SI es divisible entre el divisor, el residuo es 0
# Nota el modulo SIEMPRE te devuelve un entero.
print(8 % 2) 
print(1024 % 2)
print(9 % 3)
print(1533 % 3)
# Si el cosiente NO es divisible entre el divisor, el residuo es distinto de 0
print(3 % 2)
print(5 % 2)
print(412442672 % 3)
print(16 % 3)
```

```python
# Potencias
P1,P2 = 7,2
print(P1 ** P2)
P1,P2 = 7,8
print(P1 ** P2)
```

# Operaciones String

```python
# Nota: Al conjunto de caracteres se le denomina String.
HM = "Hola Mundo"
print(HM) # Primer codigo que todos los programadores solemos hacer.
# Tambien podemos imprimir direcatamente sin declarar variables
print("Arriba NSQK!")
# Podemos concatenar (unir) codigos.
a,b = "ta","co"
print(a + b)
# si queremos agregar un espacio es:
print(a + " " + b)
# podemos hacer mayusculas o minusculas
a,b = "man","go"
print((a+b).upper())
print("FRESA".lower())

# Tambien podemos multiplicar Strings
print('rosa' * 3)
print("multi " * 10) # se usa? la verdad no.

# La versión mas usada del print por comodidad en cuanto a Strings es:
print(f"{HM}, mi primer Hola mundo, {(a+b).upper()} ")
```
