# Condicionales


```python
# los condicionales sirven para evaluar operadores logicos
# Supongamos que estas comprando un cafe, pero si lo pides con leche deslactosada
# cuesta 5 pesos mas
cafe  = 50
leche_deslactosada = True
if leche_deslactosada: 
    cafe += 5
print(f"A pagar: ${cafe}")
```

```python
# Tambien se puede aplciar operadores dentro de la condición
promedio = 70.
print(promedio)
# En esta ocasión solo se acepta escala del 1 - 10 por lo que 70 está fuera del rango
if (promedio/10) > 7:
    print("Pasaste 1")
    
if (promedio/10) >= 7:
    print("Pasaste 2")
```

### Elif

Ahora hagamos un programa en el que preguntamos la calificación del 0 al 100 y queremos saber su equivalente en el sistema de
evaluación estadounidense. Para simplicidad, basemonos en esto:
A+ = 100
A = 90-99
B = 80-89
C = 70-79
D = 60-69
F = <60

```python
cal = 80
if cal == 100: #importante notar el uso del == para comparar si dos valores son iguales, el operador = es igual a asignacion
    print("A+")
elif cal >= 90:
    print("A")
elif cal >= 80:
    print("B")
elif cal >= 70:
    print("C")
elif cal >= 60:
    print("D")
else: #notemos como, al ya no haber una condicion, el else cubre todo lo demas
    print("F")
```

# OPERADORES LÓGICOS

Tenemos 3 tipos de operadores lógicos, los cuales nos sirven para conectar varias condiciones dentro de una misma línea:
- and
- or
- not
Veamos ahora ejemplos de sus usos.

```python
# Mismo codigo que el de arriba pero agregando una condicion mas

calificacion = 80
if calificacion >= 0 and calificacion <= 100: #notemos como usamos el and para unir dos condiciones
    if calificacion == 100:
        print("A+")
    elif calificacion >= 90:
        print("A")
    elif calificacion >= 80:
        print("B")
    elif calificacion >= 70:
        print("C")
    elif calificacion >= 60:
        print("D")
    else:
        print("F")
else:
    print("Error: La calificacion debe estar entre 0 y 100.")
    
```

```python
#El módulo es un operador matemático que devuelve el residuo de una división.
#Por ejemplo:
#10 % 3 da 1, porque 10 dividido entre 3 cabe 3 veces y sobra 1.
#15 % 5 da 0, porque 15 es múltiplo exacto de 5 y no deja residuo.
#7 % 2 da 1, porque 7 no es divisible de forma exacta entre 2.

num = 9
if num % 3 == 0 or num % 5 == 0: #notemos como usamos el or para unir dos condiciones
    print(f"El numero {num} es multiplo de 3 o de 5.")
else:
    print(f"El numero {num} no es multiplo de 3 ni de 5.")

#Sigamos con el tema de múltiplos de números para un ejemplo del operador not. Supongamos que queremos saber si un número NO 
#es par.

num = 6
if not num % 2 == 0: #notemos como usamos el not para negar la condicion
    print(f"El numero {num} no es par.")
else:
    print(f"El numero {num} es par.")
```

Los operadores lógicos son de mucha utilidad, vimos ejemplos sencillos pues los aplicaremos en ejemplos más completos en temas posteriores.

# CICLOS FOR

Los llamados ciclos 'for' nos ayudan a realizar acciones (un bloque determinado de código) durante una cantidad determinada de veces. En otras palabras, es un bucle que recorre o itera sobre los elementos de algún objeto.

```python
# Imprimimos los numeros del 0 al 4
for i in range(5): # toma 5 como el numero final pero inicia SIEMPRE en 0
    print(i)
```

```python
for num in range(1, 11):  # inicia en 1 y termina en 10
    print(num)
```

```python
#Otro tipo de For: iterar en una string
cadena = "Python"
for letra_por_letra in cadena:
    print(letra_por_letra)
#notemos como en cada iteracion, la variable letra toma el valor de cada caracter en la cadena
```

```python
#Suma de numeros: sumemos los números del 1 al número n ingresado por el usuario
n = 10
suma = 0
for i in range(1, n + 1): #range(1, n+1) genera los numeros del 1 al n inclusive
    suma += i #nueva forma de sumar, es equivalente a suma = suma + 1
print(f"La suma de los numeros del 1 al {n} es: {suma}")
```

```python
#Tambien existe range(inicio, fin, paso)
#con pasos nos referimos a cada cuanto queremos que aumente el valor, por ejemplo:

print("Numeros pares del 0 al 20: ")
for i in range(2, 21, 2):
    print(i)
```

```python
#Veamos otro ejemplo del ciclo for: contar las vocales en una cadena ingresada por el usuario
cadena = "Hola como estas"
vocales = "aeiou"
cantidad_vocales = 0

for letra in cadena: # si aqui es H...
    if letra in vocales: #aca tambien y letras in vocales busca h en la variable vocales
        cantidad_vocales += 1
print(f"Cantidad de Vocales: {cantidad_vocales}")
```

# CICLOS WHILE

```python
suma = 0 # como usamos esta misma variable arriba hay que igualarla a 0 ya que tiene el contenido anterior
numero = int(input("Ingresa un número para sumar (0 para salir): "))

while numero != 0:
    suma += numero
    numero = int(input("Ingresa otro número (0 para salir): "))

print(f"Tu suma es: {suma}")
```

```python
numero = 1
while numero <= 12:
    print(numero)
    numero += 1 #IMPORTANTE: Si no aumentamos el valor de numero, el ciclo nunca terminará
```

Vamos a hacer un pequeño juego de adivinar un numero

```python
print("Juego: Adivina el Número")
numero_secreto = int(input("Jugador 1, escribe el número secreto (1-20): "))

intentos = 0
max_intentos = 4

while intentos < max_intentos:
    intento = int(input("Adivina el número: "))
    intentos += 1

    if intento == numero_secreto:
        print("¡Correcto!")
        break
    else:
        print("Incorrecto, intenta de nuevo.")

#Opcional
if intentos == max_intentos and intento != numero_secreto:
    print("Se acabaron los intentos.")
    print("El número era:", numero_secreto)
```

Ahora vamos a adivinar una contraseña

```python
print("Sistema de Inicio de Sesión")

contraseña_correcta = "python123"
intentos = 0
max_intentos = 3

while intentos < max_intentos:
    contraseña = input("Ingresa la contraseña: ")
    intentos += 1

    if contraseña == contraseña_correcta:
        print("¡Bienvenido! Contraseña correcta.")
        break
    else:
        print("Contraseña incorrecta. Intento", intentos, "de", max_intentos)

# Mensaje final si no logró entrar
if intentos == max_intentos and contraseña != contraseña_correcta:
    print("Acceso bloqueado. Demasiados intentos.")
```

```python
#Ingresar tus calificaciones
#NOTA: Para términos prácticos, solo validaremos la entrada del usuario.
calificacion = int(input("Ingresa tu calificacion (0-100): "))
while calificacion < 0 or calificacion > 100:
    print("Error: La calificacion debe estar entre 0 y 100.")
    calificacion = int(input("Ingresa tu calificacion (0-100): "))
    
```

Del código anterior podemos apreciar algo: repetimos la línea de código que pide calificación, y si hay algo que en programación no es la mejor práctica es repetir código al menos que sea la mejor opción. ¿Qué herramienta vista podemos usar para corregir eso? Exacto, el ciclo while True (en otro lenguajes se le llama do-while, pero tiene una sintaxis diferentes, lo mencionamos en caso que buscando en línea te encuentres con dicho concepto sepas que el equivalente de do-while en Python es while True)

```python
while True:
    calificacion = int(input("Ingresa tu calificacion (0-100): "))
    if calificacion < 0 or calificacion > 100:
        print("Error: La calificacion debe estar entre 0 y 100.")
    else:
        break
```

#### EJERCICIO

- Pregunta la cantidad de alumnos y materias.

- Por cada alumno pide matrícula y calificaciones.

- Valida que matrícula y calificaciones sean correctas.

- Calcula e imprime matrícula + promedio del alumno.

- Usa los promedios individuales para calcular el promedio grupal.

- Calcula promedios por alumno y promedio grupal.

```python
#puedes usar el .strip() el cual elimina los espacios en blanco al inicio y al final de un texto.
alumnos = int(input("Alumnos: "))
materias = int(input("Materias: "))
suma_prom = 0

for i in range(alumnos):
    mat = input("Matrícula: ")
    
    suma_cal = 0
    for m in range(materias):
        cal = float(input(f"Cal {m+1}: "))
        while cal < 0 or cal > 100:
            cal = float(input("Cal válida (0-100): "))
        suma_cal += cal

    prom = suma_cal / materias
    suma_prom += prom
    print(f"Promedio de: {mat} = {prom:.2f}")#.2f es para poner 2 decimales solamente

print("Promedio grupal:", suma_prom / alumnos)
```
