## 1.- ¿Qué es POO y por qué usarla?
1. Clases (Los Moldes)
Una Clase es como el molde o el plano para crear algo.

Ejemplo: La clase "Carro".

Define qué debe tener cualquier carro (color, marca, modelo).

Define qué puede hacer cualquier carro (acelerar, frenar, encender).

2. Objetos (Las Creaciones)
   
Un Objeto es una instancia real y tangible creada a partir de una Clase.

Ejemplo:

Objeto 1: Mi carro rojo, marca Ferrari, modelo F40.

Objeto 2: Tu carro azul, marca Toyota, modelo Corolla.

Ambos son Objetos de la misma clase "Carro", pero tienen valores diferentes (rojo vs. azul, Ferrari vs. Toyota).

## 5.- Métodos: de instancia, clase

Un método de instancia es una función definida dentro de una clase que:

- Opera sobre un objeto concreto (una instancia)

- Puede acceder y modificar los atributos de ese objeto

- Recibe como primer parámetro a self

```python
# (ya lo hemos usado)
class tuClase:
    def metodo(self):
        pass
        
#metodo -> método de instancia
#self -> referencia al objeto que llama al método
```

```python
# haremos una Calculadora:
class Calculadora:
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2
    def sumar(self):
        print(self.num1 + self.num2)
    def restar(self):
        print(self.num1 - self.num2)
    def multiplicar(self):
        print(self.num1 * self.num2)
    def dividir(self):
        if self.num2 == 0:
            print("Numero no divisible")
        else:
            print(self.num1/self.num2)
            
Calculadora1 = Calculadora(7,0)
Calculadora1.dividir()
Calculadora2 = Calculadora(10,3)
Calculadora2.restar()
Calculadora2.multiplicar()
```

## 6.- Encapsulamiento y propiedades (@property)

```python
# Atributos Privados.
# Se declara con doble guión bajo
class Cuenta:
    def __init__(self, saldo):
        self.__saldo = saldo  #privado
#sirve para proteger los datos internos de un objeto y controlar cómo se leen o modifican.
cuenta = Cuenta(1000)
cuenta.Cuenta = 2000 #si puedes editarlos
#print(cuenta.__saldo)   #Error: no se puede acceder directamente


#Un microondas no te deja modificar directamente el voltaje interno. Tú solo eliges el tiempo y la potencia mediante botones, y el aparato se encarga de que todo funcione de forma segura.
#Eso es encapsulamiento: no ves ni tocas lo peligroso, solo usas la interfaz segura.
```

```python
# GETTERS Y SETTERS
class Cuenta:
    def __init__(self, saldo):
        self.__saldo = saldo #declaracion del atributo encapsulado

    @property
    def saldo(self):#GETTER
        #Permite leer el valor de __saldo usando c.saldo
        return self.__saldo

    @saldo.setter # NOTA: si se declara el property obligatoriamente debes declarar el setter
    def saldo(self, valor):#SETTER
        #Se ejecuta cuando hacemos: c.saldo = valor
        self.__saldo = valor


c = Cuenta(300)
print(c.saldo)
c.saldo = 4000
print(c.saldo)
c.saldo
```

## 7.- Herencia y Abstracion

```python
#Herencia
class Vehiculo: #Classe base (a mi me gusta decirle clase padre)
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        return f"{self.marca} {self.modelo}"

    def mover(self):
        return "El vehículo se está moviendo..."
# ----------------------------------------------------------        
class Coche(Vehiculo): #subclase 1 (Clase hija)
    def mover(self):
        return "El coche está conduciendo por la carretera" #espera, puedes editar el contenido de algún método heredado? Si, lo veremos en un rato
#-----------------------------------------------------------        
class Moto(Vehiculo): #subclase 2 (Clase hija)
    def mover(self):
        return "La moto avanza sobre dos ruedas"

coche2 = Vehiculo("BYD", "Dolphin")
coche = Coche("Toyota", "Corolla")
moto = Moto("Yamaha", "R6")

print(coche2.descripcion())
print(coche2.mover())

print(coche.descripcion()) #aun que no volvimos a declarar el metodo dentro de descripcion, podemos seguir usandolo ya que lo heredo del padre
print(coche.mover())

print(moto.descripcion())
print(moto.mover())
#En pocas palabras puedes editar el contenido de un metodo o puedes dejarlo asi (pero se quedara como el metodo original[del padre])
```

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        print(f"Hola, soy {self.nombre} y tengo {self.edad} años.")


class Estudiante(Persona):
    def __init__(self, nombre, edad, carrera):
        super().__init__(nombre, edad)
        self.carrera = carrera

    def presentarse(self):
        super().presentarse()
        print(f"Estudio la carrera de {self.carrera}.")


#Reutiliza el constructor de la clase padre (lo cual evita duplicar código)
#Ejecuta el comportamiento base y luego lo extiende
est = Estudiante("Ana", 21, "Ingeniería")
est.presentarse()
```

## Composicion HAS-A
¿Por qué se llama composición?

Porque estás componiendo (armando) un objeto grande usando otros objetos.
Es un principio de programación orientada a objetos donde:

Una clase NO hereda de otra

Sino que contiene un objeto de esa otra clase


```python
class Banco:
    def __init__(self, dinero):
        self.dinero = dinero

    def ver_total(self):
        print(f"Total en la cuenta: ${self.dinero}")

    def retirar(self, monto):
        if monto <= self.dinero:
            self.dinero -= monto
            print(f"Retiraste ${monto} con éxito")
        else:
            print("Fondos insuficientes")

    def retiro_sintarjeta(self, monto):
        if monto <= self.dinero:
            self.dinero -= monto
            print(f"Retiraste ${monto} sin tarjeta con éxito")
        else:
            print("Fondos insuficientes para retiro sin tarjeta")
            
class Persona:
    def __init__(self, cuenta: Banco): #indica que este parámetro debería ser una instancia de la clase Banco.
        self.cuenta = cuenta   # Persona TIENE un Banco, no hereda de él

    def retirar_sin_tarjeta(self, monto):
        self.cuenta.retiro_sintarjeta(monto)
        
cuenta = Banco(5000)
persona = Persona(cuenta)

persona.retirar_sin_tarjeta(500)
cuenta.ver_total()
```

## 9.-Composición vs herencia

Esto se va a responder con pura teoria: 
No significa que alguna sea mejor que otra simplemente que aveces una te puede conveniar más que otra:
Herencia:
Un gato ES un animal
(no "Un gato TIENE un aniaml)"

Composición:
Una persona TIENE una Cuenta
(no "Una Persona ES un Banco”).

```python
'''
Ejercicio Aplicando todo:

Queremos modelar un pequeño sistema para vehículos.
Habrá distintos tipos:
Coche
Moto
Bicicleta

Todos son “vehículos” (clase vehiculo), por lo tanto comparten características básicas, pero cada uno se mueve distinto, 
debes poner una clase llamada motor la cual use Composición, opcional: Usa Polimorfismo
'''
class Vehiculo:
    def __init__(self, marca):
        self.marca = marca

    def mover(self):
        raise NotImplementedError("Este método DEBE ser implementado.")


class Motor:
    def __init__(self, caballos):
        self.caballos = caballos

    def potencia(self):
        return f"{self.caballos} HP"


class Coche(Vehiculo):
    def __init__(self, marca, caballos): 
        super().__init__(marca)
        self.motor = Motor(caballos)   # COMPOSICIÓN

    def mover(self): #POLIMORFISMO
        return f"El coche {self.marca} avanza con motor de {self.motor.potencia()}"


class Moto(Vehiculo): #POLIMORFISMO
    def __init__(self, marca, caballos):
        super().__init__(marca)
        self.motor = Motor(caballos)   # COMPOSICIÓN

    def mover(self):
        return f"La moto {self.marca} acelera con motor de {self.motor.potencia()}"


class Bicicleta(Vehiculo):
    def mover(self): #POLIMORFISMO
        return f"La bicicleta {self.marca} avanza con pedaleo humano"


def probar_vehiculos(lista):# --- llamada al polimorfismo ---
    for v in lista:
        print(v.mover())


#lo usaremos en esta ocasion en una lista, SI se puede hacer listas de clases.
vehiculos = [
    Coche("Toyota", 150),
    Moto("Yamaha", 90),
    Bicicleta("Trek")
]

probar_vehiculos(vehiculos)
```
