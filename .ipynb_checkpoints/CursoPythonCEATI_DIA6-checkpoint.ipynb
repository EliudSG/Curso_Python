## 1.- ¿Qué es POO y por qué usarla?
1. Clases (Los Moldes)
Una Clase es como el molde o el plano para crear algo.

Ejemplo: La clase "Carro".

Define qué debe tener cualquier carro (color, marca, modelo).

Define qué puede hacer cualquier carro (acelerar, frenar, encender).

2. Objetos (Las Creaciones)
Un Objeto es una instancia real y tangible creada a partir de una Clase.

Ejemplo:

Objeto 1: Mi carro rojo, marca Ferrari, modelo F40.

Objeto 2: Tu carro azul, marca Toyota, modelo Corolla.

Ambos son Objetos de la misma clase "Carro", pero tienen valores diferentes (rojo vs. azul, Ferrari vs. Toyota).

```python
class Perro:
    especie = "Canis Familiaris"  # atributo de clase
    patas = 4
    def __init__(self, nombre):
        self.nombre = nombre      # atributo de instancia

p1 = Perro("Firulais")
p2 = Perro("Bobby")

print(p1.especie) 
print(p2.especie) 
# SE PUEDE MODIFICAR? RESPUESTA: SI
print("-----------------")
p2.especie = "Gato"
print(p1.especie) #cambió el atributo de la clase ESPECIALMENTE para p1
print(p2.especie)
```

## 5.- Métodos: de instancia, clase y estáticos

```python
# Calculadora:
class Calculadora:
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2
    def sumar(self):
        print(self.num1 + self.num2)
    def restar(self):
        print(self.num1 - self.num2)
    def multiplicar(self):
        print(self.num1 * self.num2)
    def dividir(self):
        if self.num2 == 0:
            print("Numero no divisible")
        else:
            print(self.num1/self.num2)
            
Calculadora1 = Calculadora(7,0)
Calculadora1.dividir()
Calculadora2 = Calculadora(10,3)
Calculadora2.restar()
Calculadora2.multiplicar()
```

## 6.- Encapsulamiento y propiedades (@property)

```python
# Atributos Privados.
# Se declara con doble guión bajo
class Cuenta:
    def __init__(self, saldo):
        self.__saldo = saldo  #privado

cuenta = Cuenta(1000)
cuenta.Cuenta = 2000 #si puedes editarlos
#print(cuenta.__saldo)   #Error: no se puede acceder directamente
```

```python
# Para poder acceder a ellos:
class Cuenta:
    def __init__(self, saldo):
        self.__saldo = saldo

    def obtener_saldo(self):
        return self.__saldo

    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto

    def retirar(self, monto):
        if 0 < monto <= self.__saldo:
            self.__saldo -= monto
c = Cuenta(2000)
TU_SALDO = c.obtener_saldo()
print(TU_SALDO)
```

```python
# GETTERS Y SETTERS
class Cuenta:
    def __init__(self, saldo):
        self.__saldo = saldo

    @property
    def saldo(self):              #getter
        return self.__saldo

    @saldo.setter
    def saldo(self, valor):       #setter
        if valor >= 0:
            self.__saldo = valor
        else:
            raise ValueError("El saldo no puede ser negativo")
c = Cuenta(300)
print(c.saldo)    # obtener (getter)
c.saldo = 4000    # asignar (setter)
print(c.saldo)
c.saldo
```

## 7.- Herencia y especialización

```python
class Vehiculo: #Classe base
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        return f"{self.marca} {self.modelo}"

    def mover(self):
        return "El vehículo se está moviendo..."
# ----------------------------------------------------------        
class Coche(Vehiculo): #subclase 1 (Clase hija)
    def mover(self):
        return "El coche está conduciendo por la carretera"
#-----------------------------------------------------------        
class Moto(Vehiculo): #subclase 2 (Clase hija)
    def mover(self):
        return "La moto avanza sobre dos ruedas"
        
coche2 = Vehiculo("BYD", "Dolphin")
coche = Coche("Toyota", "Corolla")
moto = Moto("Yamaha", "R6")

print(coche2.descripcion())
print(coche2.mover())

print(coche.descripcion())
print(coche.mover())

print(moto.descripcion())
print(moto.mover())
```

```python
class Animal:#Animal es una clase base abstracta: no se puede usar directamente porque hablar() no está implementado.
    def hablar(self): 
        raise NotImplementedError

class Gato(Animal): #Gato heredó todo lo que Animal tiene, ya que el gato es un animal
    def hablar(self):
        return "miau"


class Perro(Animal):
    def hablar(self):
        return "guau"

class Pajaro(Animal):
    def hablar(self):
        return "piopio"
        
Naylah = Perro()
print(Naylah.hablar()) #polimorfismo en acción
```

## Composicion HAS-A

```python
class Banco:
    def __init__(self, dinero):
        self.dinero = dinero

    def ver_total(self):
        print(f"Total en la cuenta: ${self.dinero}")

    def retirar(self, monto):
        if monto <= self.dinero:
            self.dinero -= monto
            print(f"Retiraste ${monto} con éxito")
        else:
            print("Fondos insuficientes")

    def retiro_sintarjeta(self, monto):
        if monto <= self.dinero:
            self.dinero -= monto
            print(f"Retiraste ${monto} sin tarjeta con éxito")
        else:
            print("Fondos insuficientes para retiro sin tarjeta")
            
class Persona:
    def __init__(self, cuenta: Banco): #indica que este parámetro debería ser una instancia de la clase Banco.
        self.cuenta = cuenta   # Persona TIENE un Banco, no hereda de él

    def retirar_sin_tarjeta(self, monto):
        self.cuenta.retiro_sintarjeta(monto)
        
cuenta = Banco(5000)
persona = Persona(cuenta)

persona.retirar_sin_tarjeta(500)
cuenta.ver_total()
```

## 9.-Composición vs herencia

Ejercicio Aplicando todo:

Queremos modelar un pequeño sistema para vehículos.
Habrá distintos tipos:
Coche
Moto
Bicicleta

Todos son “vehículos” (clase vehiculo), por lo tanto comparten características básicas, pero cada uno se mueve distinto, 
debes poner una clase llamada motor la cual use Composición, opcional: Usa Polimorfismo

```python
class Vehiculo:
    def __init__(self, marca):
        self.marca = marca

    def mover(self):
        raise NotImplementedError("Este método DEBE ser implementado.")


class Motor:
    def __init__(self, caballos):
        self.caballos = caballos

    def potencia(self):
        return f"{self.caballos} HP"


class Coche(Vehiculo):
    def __init__(self, marca, caballos): 
        super().__init__(marca)
        self.motor = Motor(caballos)   # COMPOSICIÓN

    def mover(self): #POLIMORFISMO
        return f"El coche {self.marca} avanza con motor de {self.motor.potencia()}"


class Moto(Vehiculo): #POLIMORFISMO
    def __init__(self, marca, caballos):
        super().__init__(marca)
        self.motor = Motor(caballos)   # COMPOSICIÓN

    def mover(self):
        return f"La moto {self.marca} acelera con motor de {self.motor.potencia()}"


class Bicicleta(Vehiculo):
    def mover(self): #POLIMORFISMO
        return f"La bicicleta {self.marca} avanza con pedaleo humano"


def probar_vehiculos(lista):# --- llamada al polimorfismo ---
    for v in lista:
        print(v.mover())


#lo usaremos en esta ocasion en una lista, SI se puede hacer listas de clases.
vehiculos = [
    Coche("Toyota", 150),
    Moto("Yamaha", 90),
    Bicicleta("Trek")
]

probar_vehiculos(vehiculos)
```
