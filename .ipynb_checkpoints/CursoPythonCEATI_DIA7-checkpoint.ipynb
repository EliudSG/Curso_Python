## Limpieza de Datos

```python
df_clean = df.copy()
print(df_clean)
# Rellenar valores faltantes
df_clean["edad"] = df_clean["edad"].fillna(df_clean["edad"].mean())     # numérico → se usa la media
df_clean["ingresos"] = df_clean["ingresos"].fillna("$0")                 # texto → marcador claro
df_clean["nombre"] = df_clean["nombre"].fillna("Desconocido")            # texto → etiqueta clara

# Explicación:
# fillna() = rellena valores faltantes

# https://pandas.pydata.org/docs/reference/frame.html?utm_source=
# mean()   = calcula la media de la columna (solo numérico)
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html?utm_source=

# Por qué:
# - edad: la media conserva el rango natural de la variable sin sesgar hacia extremos.
# - ingresos: "$0" funciona como indicador de "sin registro".
# - nombre: "Desconocido" es mejor que eliminar filas completas.

df_clean
```

```python
# Comparación de DataFrame sin limpiar vs limpio
print(df)
print(df_clean)
```

```python
# Ahora acordemonos que los estados están escritos mal, algunos en mayusculas, otros en minusculas
df_clean["ciudad"] = (
    df_clean["ciudad"]
    .str.lower()      # todo en minúsculas o upper para mayusculas
    .str.strip()       # quitar espacios
)
df_clean
```

## Introduccion a la regresión lineal

```python
from sklearn.linear_model import LinearRegression
```

# ¿Qué es una regresión?
Una regresión intenta predecir un valor numérico a partir de otros datos.

Ejemplos:
- Predecir ingresos según edad.
- Predecir precio según tamaño.

Aquí usaremos regresión lineal: el tipo más simple y común.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=df_clean["edad"], y=df_clean["ingresos"])
plt.xlabel("Edad")
plt.ylabel("Ingresos")
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

```python
# Todas las regresiones lineales tienen variables de entrada (puedes tener MULTIPLES) y de salida (usualmente de 1)
X = df_clean[["edad"]]      # variable de entrada
y = df_clean["ingresos"]    # lo que queremos predecir

modelo = LinearRegression()
modelo.fit(X, y)
```

Tiene relacion con la formula de la pendiente.
¿Qué hace fit?
Busca los valores de m (pendiente) y b (intercepto) que mejor ajustan tus datos.
Minimiza el error entre los ingresos reales y los ingresos predichos.
Después de fit, el modelo ya “aprendió” la relación:
ingresos≈b⋅edad + a 
formula: mx + b

```python
edad_nueva = [[30]]
prediccion = modelo.predict(edad_nueva)
prediccion
#dara un Warning “Me entrenaste con features que tenían nombres,
#pero ahora me estás dando datos sin nombres. Puedo seguir, pero te aviso que puede ser inconsistente.” 
```

### Interpretación sencilla
El modelo aprendió la relación entre:
- edad
- ingresos

Y ahora puede estimar ingresos de alguien con edad que no estaba en el dataset.
