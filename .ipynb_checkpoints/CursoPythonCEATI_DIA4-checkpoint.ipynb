# FUNCIONES

En este curso hemos usado algunas funciones integradas en Python, pero también nosotros podemos definir funciones, siendo estas un bloque de código reutilizable que podemos mandar a llamar varias veces en el código, estas pueden o no regresar algo.
Empezemos con un ejemplo que no regresa nada.

```python
#Esta función simplemente imprime un mensaje, observemos como sus paréntesis están vacíos

def mensaje():
    print("Hola!!!")

print("Mi mensaje es: ")
mensaje() #esta es la forma en la que SIEMPRE mandaremos a llamar la funcion
```

```python
# Quiero que ustedes vean que también se puede retornar variables (de cualquier tipo de datos)
def mensaje():
    return "Hola!!!" #puede retornar False y True, numeros, flotantes literalmente TODO
print(mensaje())
```

```python
#Ahora, en los paréntesis podemos tener parámetros, por ejemplo, la función anterior podría recibir un nombre

def saludo(nombre):
    print("Hola ", nombre, "!!!") #Aqui lo estamos imprimiendo

tuNombre = input("Ingresa tu nombre: ")
saludo(tuNombre)
#NOTA: El nombre de la variable del parámetro puede o no ser el mismo fuera de la función
```

```python
#También podemos definir el parámetro que enviaremos al usar la función
saludo("grupo")
```

```python
# Aqui mismo otra funcion pero con numeros
def suma(numero_1, numero_2) : # podemos ponerle tantos parametros como queramos
    return numero_1 + numero_2
    
suma = suma(1,5)
print(suma)
```

```python
#Ahora al definir las funciones podemos definir un parámetro por defecto en caso
def bienvenida(nombre="invitado"):
    print("Hola", nombre)

bienvenida()
bienvenida("Victor")
```

```python
# Aqui mismo otra funcion pero con numeros
def suma(numero_1 = 2, numero_2 = 10) : # podemos ponerle tantos parametros como queramos
    return numero_1 + numero_2
    
suma = suma(1000000,1512350)
print(suma)
```

```python
#DESCUENTO
#Claramente podemos tener más de dos parámetros, por ejemplo, si queremos calcular un descuento
#Supongamos que el descuento usual es de 15% al menos que se indique lo contrario
def descuento(precio, desc=0.15):
    precio = precio * (1-desc)
    print("El nuevo precio es", precio)

descuento(750)
descuento(800, 0.5)
#Observa como primero se coloca el parámetro que no tiene un valor default para poder mandarlo correctamente al llamar a
#la función
```

Podemos tener muchas funciones en un código que pueden recibir gran variedad de parámetros y regresar más de un solo resultado. Antes de avanzar con ejemplo más completos de funciones veremos el siguiente tema del día.

```python
# Las funciones que retornan mas de 1 valores los retorna como tuplas
def usuario(nombre, edad):
    return (nombre, edad)

resultado = usuario("Ana", 25)
print(type(resultado))
print(resultado)
```

```python
# Otro ejercicio donde la funcion regrese una TUPLA
def operaciones(numero):
    doble = numero * 2
    triple = numero * 3
    cuadruple = numero * 4
    quintuple = numero * 5
    return (doble, triple, cuadruple, quintuple)

resultado = operaciones(2)
print(resultado)
```

# Funciones que llaman a otras Funciones

```python
def saludar(nombre):
    return f"Hola, {nombre}!"

def despedir(nombre):
    return f"Adiós, {nombre}!"

def mensaje_completo(nombre):
    # Esta función llama a las otras dos
    saludo = saludar(nombre)
    despedida = despedir(nombre)
    return saludo + " " + despedida # el " " son para tener un espacio ya que si no se encontraria un poco pegado

print(mensaje_completo("Ana"))
```

Ahora hagamos un ejercicio super rapido de una heladeria

```python
# Opcional
def pedir_sabor():
    sabores = ["vainilla", "chocolate", "fresa"]
    print("Sabores disponibles:", ", ", sabores)
    sabor = input("Elige un sabor: ").strip().lower()
    while sabor not in sabores:
        sabor = input("Error, elige un sabor válido: ").strip().lower()
    return sabor

def pedir_cantidad():
    cantidad = int(input("¿Cuántas bolas quieres? "))
    while cantidad <= 0:
        cantidad = int(input("Error, ingresa un número mayor a 0: "))
    return cantidad

def pedir_helado():
    sabor = pedir_sabor()   # llama a pedir_sabor
    cant = pedir_cantidad() # llama a pedir_cantidad
    print(f"Pedido: {cant} bola(s) de {sabor}")
    return cant, sabor

total_pedidos = int(input("¿Cuántos clientes atenderás? "))
for _ in range(total_pedidos):
    pedir_helado()
```

# FUNCIONES LAMBDA

Una función lambda es una función acortada para escribir una función de esta forma:
nombreFuncion = lambda parámetros: acción
Algo importante a saber es que no es necesario definir su nombre, sin embargo, para que sean más útiles es necesario.

```python
#Veamos un ejemplo de definir la función de sumar
sumar = lambda x,y: x+y
print(sumar(3,10))
```

```python
#Realmente tienen muchas limitaciones frente a las funciones normales, pero pueden ser útiles para funciones muy simples
#También pueden devolver más de un valor
descuentoLambda = lambda desc, precio: ((precio*(1-desc)), (precio*desc))
nuevoPrecio = descuentoLambda(0.4, 500)
print("El precio a pagar es", nuevoPrecio[0], ",el descuento que se hizo fue de", nuevoPrecio[1])
```

# ERRORES

Probablemente ya te has preguntado cómo manejar ciertos errores, como que el usuario ingrese cierto de dato, que de equivocarse el programa no 'truene' sino que vuelva a pedir el dato y siga intentando. Para ello podemos usar los errores definidos en Python con bloques try/except

```python
#Empezemos con un ejemplo muy clásico: la división entre 0. Supongamos que tenemos este código
dividendo = 10
divisor = 0
division = dividendo/divisor
print(division)
print("No llega a esto por el error") #ni siquiera imprime ya esto
#Observa que el programa marca el eeror de ZeroDivisionError
#Aquí lo definimos para practicidad, el caso sería que el usuario lo ingrese directamente
#Esto directamente hace que el programa deje de funcionar
```

```python
#Hagmos esto en un bloque try/except, ahora si pedimos los datos
try:
    dividendo = float(input("Ingrese el dividendo: "))
    divisor = float(input("Ingrese el divisor: "))
    division = dividendo/divisor
    print(division)
except ZeroDivisionError:
    print("Error: el divisor no debe ser 0")

print("No llega a esto por el error") #ahora incluso si hay un 0, si imprime esto
```

```python
#Tenemos otros tipos de errores, por ejemplo, el de ValueError que sale al ingresar el tipo incorrecto de dato
#Digamos que un usuario ingresa texto en un lugar de un número, esto nos ayuda a manejarlo
try:
    numero = int(input("Escriba un número: "))
    print(numero)
except ValueError:
    print("Ingrese un valor numérico")
```

```python
#También podemos manejar más de una excepción, por ejemplo, con la división podemos tener tanto el de tipo de valor como el 
#de la división entre 0
try:
    dividendo = float(input("Ingrese el dividendo: "))
    divisor = float(input("Ingrese el divisor: "))
    division = dividendo/divisor
    print(division)
except ZeroDivisionError:
    print("Error: el divisor no debe ser 0")
except ValueError:
    print("Ingrese un valor numérico")
```

```python
#Con esto ya visto podemos regresar más ejemplos de funciones
#Primer ejemplo: Una función que recibe una lista y obtiene el promedio del grupo y el mayor promedio
#NOTA: Solo consideraremos cuál es el mayor, no cuántas veces se pueda repetir ese promedio
def promediosYmayor(promedios):
    promedio = 0
    for x in promedios:
        promedio += x
    promedio = promedio/len(promedios)
    return promedio, max(promedios)

promedios = []
continuar = ""

while True:
    if len(promedios) > 0:
        continuar=""
        while continuar not in ("si", "no"):
            continuar = input("Ingrese una respuesta válida: si/no").lower()
        if continuar == "no":
            break

    while True:
        try:
            calif = int(input("Indique la calificación: "))
            if calif < 0 or calif > 100:
                print("Ingrese un valor valido")
            else:
                promedios.append(calif)
                break
        except ValueError:
            print("Ingrese un valor numérico")

prom, mayor = promediosYmayor(promedios)

print("\nResultados:")
print("Promedio:", prom)
print("Mayor:", mayor)
```

```python
#Ahora hagamos un ejemplo más completo: una calculadora con suma, resta, producto, división y potencia.

#Primero definimos las funciones
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def producto(a, b):
    return a * b

def division(a, b):
    if b == 0: #revisar la división entre 0
        return "Error: no se puede dividir entre 0"
    return a / b

def potencia(a, b):
    return a ** b


#Para mostrar el menú
#Las funciones son útiles para que el main o flujo principal del trabajo sea fácil de revisar, es decir
#que no este saturado de código sino que se trabaje en partes por funciones y esto es un ejemplo.
def menu():
    print("MI CALCULADORA")
    print("1) Suma")
    print("2) Resta")
    print("3) Producto")
    print("4) División")
    print("5) Potencia")
    print("6) Salir")


while True:
    menu()
    try:
        op = int(input("Seleccione una opción: "))
        if op < 1 or op > 6:
            print("Opción inválida. Elija del 1 al 6.")
            continue
    except ValueError:
        print("Ingrese un número válido.")
        continue

    if op == 6:
        print("SALIENDO....")
        break

    #Pedir los datos
    try:
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
    except ValueError:
        print("Error: ingrese solo números.")
        continue
    #se hace la operación
    if op == 1:
        print("Resultado:", suma(a, b))
    elif op == 2:
        print("Resultado:", resta(a, b))
    elif op == 3:
        print("Resultado:", producto(a, b))
    elif op == 4:
        print("Resultado:", division(a, b))
    elif op == 5:
        print("Resultado:", potencia(a, b))
```
