## Qué es un modulo?
un módulo es una unidad de organizacion del codigo que permite encapsular y dividir el software en partes más manejables, generalmente representado como un archivo que contiene código relacionado, como funciones, variables o clases.
Existen MILES de modulos (librerias) de las cuales pondre de una manera rapida las mas usadas

```python
# Algunas librerias vienen por defecto en PYTHON, pero otras son necesarias descargarla, para esto se usa el comando
#"pip instsall libreria_que_deseas_Descargar" o "python -m pip install libreria_que_deseas_Descargar"
#Libreria Math
import math # siempre que desees importar un modulo debes de agregar import antes
# con eso ya puedes usar las funciones, variables o clases del modulo
r = math.sqrt(2) #de math, trae sqrt y sacale sqrt(raiz) de 2
area = math.pi * r**2 #de math saca PI
print(f"r={r:.4f}, area={area:.4f}")
```

```python
# tambien podemos importar ciertas funciones y/o cambiarle el nombre UNICAMENTE A NUESTRO SCRIPT
from math import pi # hay modulos que son DEMASIADO extensos, entonces esto es UTIL
print(pi) #al importar UNA funcion/clase/variables YA NO ES NECESARIO escribir la libreria antes
#renombramos math
print("---------------")
import math as matematicas
print(matematicas.sqrt(2))
print("----------------")
# o tambien podemos hacer esto:
from math import tan as tangente
print(tangente(10))
```

### Existe 1 solo tema que TODO programador independientemente del área DEBE saber
JSON es un formato para guardar y transmitir datos.
Es muy usado en APIs, archivos de configuración y almacenamiento de información.
(Se parece a un diccionario de Python, pero es solo texto)

Para que sirve?
- Sirve para convertir objetos de Python → texto JSON y también texto JSON → objetos de Python.

Esto se llama:
- Serializar: convertir un objeto de Python en texto JSON.
- Deserializar: convertir texto JSON a un objeto usable en Python.

```python
#Funciones principales
#1) json.dumps(obj): Convierte un diccionario u objeto de Python en un string JSON

import json
obj = {'nombre': 'Ana', 'edad': 25}
print(type(obj))
s = json.dumps(obj, ensure_ascii=False)
print(s)
print(type(s)) 
```

```python
#json.loads(texto_json): Convierte un string JSON en un diccionario Python.
restaurar = json.loads(s)
print(restaurar)
#vuelve a hacer un dicc
print(type(restaurar))
# Al venir de una pagina/persona/archivo desconocid@ debes de tener cuidado
'''
Pueden contener:

Datos malformados
Información inesperada
Campos faltantes
Datos demasiado grandes
va incorrectos que rompan tu programa
'''
```

```python
'''#UTF-8 es la abreviatura de "8-bit Unicode Transformation Format" y 
se trata de un formato de codificación de caracteres Unicode e ISO 10646 
que utiliza símbolos de longitud variable, permitiendo representar cualquier 
carácter Unicode con una eficiencia que lo hace ideal para internet.'''

with open("datos.json", "w", encoding="utf-8") as f:
    json.dump(obj, f, ensure_ascii=False, indent=4)
```

```python
with open("datos.json", "r", encoding="utf-8") as f:
    # lo ideal es hacer: variable = json.load(f)
    print(json.load(f))
```

### Numpy
NumPy provee el tipo ndarray, operaciones vectorizadas, broadcasting, utilidades de álgebra lineal y funciones matemáticas en arrays. Es la base sobre la que se apoyan muchas librerías científicas de Python

#### Nota importante: El arreglo de NUMPY no es igual a la lista de python, este NO puede tener diferentes tipos de datos en el mismo

```python
#Colab ya la tiene por default, pero si no es pip install numpy
import numpy as np
#numpy tiene sus propios tipos de arreglos para poder hacer calculos con ellos
A = np.array([[1,2,3],[4,5,6]]) # array (2,3)
v = np.array([10,20,30]) # array (3,)
# puedes hacer operaciones con el arreglo de maneras en las que con las listas default de python NO puedes
print(A + v) 
print("-----------------") #vemos las diferencias
prueba = [1,2,3]
prueba2 = [1,2,"hola"]
print(prueba+prueba2) # hace una especie de union
```

Funciona la suma de esta manera:

[1, 2, 3] + [10, 20, 30] = [11, 22, 33]

[4, 5, 6] + [10, 20, 30] = [14, 25, 36]


```python
# Más funciones de Numpy
z = np.zeros((3,3)) #arreglo de ceros
print(z)
o = np.ones((2,4)) #arreglo de 1
print(o)
r = np.random.rand(3,3) #arreglo con va aleatorios
print(r)
```

```python
# Rangos Numericos
nums = np.arange(0, 10, 2) # saca los va entre 0 y 10 de 2 en 2 sin contar el 10
print(nums)
x = np.linspace(0, 1, 5) # saca 5 va espaciados de una manera "equitativa" entre 0 y 1
print(x)
```

```python
#claramente se pueden hacer operaciones con los arreglos
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print(a + b)  # suma
print(a * b)  # multiplicación
print(a ** 2) # potencia
print(b / a)  # división
```

```python
# Tambien puedes hacer el producto entre dos matrices
print(a.dot(b))
c = np.array([[1, 2, 3], [4, 5, 6]]) 
d = np.array([[10, 20], [30, 40], [50, 60]])

#1*10 + 2*30 + 3*50
#1*20 + 2*40 + 3*60
#4*10 + 5*30 + 6*50
#4*20 + 5*40 + 6*60

print(c.dot(d))
```

```python
data = np.array([1,2,3,4,5])

print(data.mean())   # promedio
print(data.sum())    # suma
print(data.min())    # mínimo
print(data.max())    # máximo
print(data.std())    # desviación estándar
```

```python
# Podemos acceder a ciertos va segun como querramos
a = np.array([10,20,30,40,50])

print(a[0])     # primer elemento
print(a[1:4])   # del índice 1 al 3
print(a[::-1])  # al revés
```

```python
# Sacar minimos y maximos de un arreglo
A = np.array([[1,5,3],
              [7,2,8]])

print(A.max(axis=0))  # por columnas
print(A.max(axis=1))  # por filas
```

```python
# Podemos poner condicionales para traer los va que querramos
A = np.array([1, 6, 3, 9, 2])
print(A[A > 5])
```

### Matplotlib (Visualizacion de datos [Graficas])

```python
# Nuevamente es una Libreria Externa asi que usamos el pip install matplotlib
import matplotlib.pyplot as plt
```

#### Grafico de Lineas

```python
#Une los graficos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
#marker='o', linestyle='-', linewidth=2
plt.plot(x, y, marker = "o") #Dibuja una línea uniendo los puntos (x,y)
plt.title("Titulo de tu grafica")#Título
plt.xlabel("Eje X") #texto horizontal
plt.ylabel("Eje Y") #Texto vertical

plt.show() 
```

#### Marker existente
| Símbolo | Nombre             | Descripción                    |
| ------- | ------------------ | ------------------------------ |
| `'o'`   | **circle**         | Un círculo                     |
| `'.'`   | **point**          | Un punto muy pequeño           |
| `','`   | **pixel**          | Un punto del tamaño de 1 píxel |
| `'x'`   | **x**              | Una X                          |
| `'+'`   | **plus**           | Un signo más                   |
| `'s'`   | **square**         | Un cuadrado                    |
| `'D'`   | **diamond**        | Un diamante grande             |
| `'d'`   | **thin diamond**   | Un diamante más delgado        |
| `'^'`   | **triangle up**    | Triángulo apuntando arriba     |
| `'v'`   | **triangle down**  | Triángulo apuntando abajo      |
| `'>'`   | **triangle right** | Triángulo derecha              |
| `'<'`   | **triangle left**  | Triángulo izquierda            |
| `'p'`   | **pentagon**       | Pentágono                      |
| `'*'`   | **star**           | Estrella                       |
| `'h'`   | **hexagon1**       | Hexágono estilo 1              |
| `'H'`   | **hexagon2**       | Hexágono estilo 2              |


#### Linestyle
| Símbolo | Nombre      | Descripción               |
| ------- | ----------- | ------------------------- |
| `'-'`   | **solid**   | Línea continua            |
| `'--'`  | **dashed**  | Línea discontinua (rayas) |
| `'-.'`  | **dashdot** | Línea punto-raya          |
| `':'`   | **dotted**  | Línea punteada            |


#### Grafico de Dispersion

```python
# Sirve para mostrar cada punto sin unirlos.
import matplotlib.pyplot as plt
import random

x = [random.randint(1,100) for _ in range(50)]
y = [random.randint(1,150) for _ in range(50)]
plt.scatter(x, y)
plt.title("Gráfico de puntos")
plt.xlabel("Número")
plt.ylabel("Número al cuadrado")

plt.show()
```

#### Grafica de Barras

```python
cat = ["A", "B", "C", "D"]
va = [5, 3, 8, 6]
plt.bar(cat, va)#hace barras verticales
plt.title("Gráfico de barras")
plt.ylabel("va")
plt.show()
```

```python
#OPCIONAL
# Usemos ahora una fusion entre matplotlib y numpy
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x) 

plt.plot(x, y, marker = "o")
plt.title("Función seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True) #activa una cuadrícula de fondo para leer mejor
plt.show()
```

#### Grafica de PIE

```python
valores = [40, 30, 20, 10] #acordemonos que la grafica de pie solo trabaja con valores la suma igual a 100
categorias = ["A", "B", "C", "D"]

plt.pie(valores, labels=categorias, autopct="%1.1f%%")
#Labels son etiquetas de cada espacio
#autopct pone el porcentaje en el espacio

plt.title("Grafica de pastel")
plt.show()
```
